```python
english_correct = ''
for i in range(len(english)):
    if english[i] not in '0123456789':
        english_correct += english[i]
```

```python
print(english_correct)
```

```python
bigrams = {}
for i in range(len(english) - 1):
    bg = english[i] + english[i + 1]
    if bg not in bigrams:
        bigrams[bg] = 1
    else:
        bigrams[bg] += 1
    
```

```python
print(bigrams)
```
